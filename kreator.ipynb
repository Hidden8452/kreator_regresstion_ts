```python


def get_split(df, columns=[], splitstr="None", n=0, col_names=[]):
    for column in columns:
        # Split by the provided delimiter (splitstr)
        split_cols = df[column].str.split(splitstr, expand=True)

        for i in range(min(n, len(split_cols.columns))):
            if i < len(col_names):
                df[col_names[i]] = split_cols[i]
    
    # Extract numeric values before and after "x" or " x "
    def extract_dimensions(value):
        # Regular expression to match formats like "510 x 240cm", "510 x 240", "960x360"
        match = re.match(r"(\d+)\s?x\s?(\d+)", str(value))
        if match:
            return int(match.group(1)), int(match.group(2))
        return None, None  # If no match, return None for both values
    
    # Create new columns with the numeric values
    df[col_names[0]], df[col_names[1]] = zip(*df[column].apply(extract_dimensions))
    
    return df
```

```python
from pandas import DataFrame

def set_dodavatel_column(df, column_name, dataframe_name):
 
    if column_name == "Dodavatel":
        df[column_name] = dataframe_name
    else:
        df[column_name] = ""

def ensure_columns_with_dtypes(dfs, columns_with_dtypes, dataframe_names=None):

    if dataframe_names is None or len(dataframe_names) != len(dfs):
        raise ValueError("dataframe_names must be provided and match the length of dfs.")

    for df, name in zip(dfs, dataframe_names):
        for col, dtype in columns_with_dtypes.items():
            if col not in df.columns:
                if dtype == "INTEGER":
                    df[col] = 0
                elif dtype == "FLOAT":
                    df[col] = 0.0
                elif dtype.startswith("VARCHAR") or dtype == "STRING":
                    set_dodavatel_column(df, col, name)  # Call the function to handle 'Dodavatel'
                else:
                    raise ValueError(f"Unsupported data type: {dtype}")
    return dfs
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, ForeignKey, JSON, Date
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship

Base = declarative_base()

class Dodavatelia(Base):
    __tablename__ = 'dodavatelia'
    id_Dodavatel = Column(Integer, primary_key=True, autoincrement=True)
    adresa = Column(String(255), nullable=True)
    telefon = Column(Integer, nullable=True)
    mail = Column(String(255), nullable=True)
    
    # Relationships
    centralny_zoznam = relationship("CentralnyZoznam", back_populates="dodavatel")

class Ponuka(Base):
    __tablename__ = 'ponuka'
    id_Ponuka = Column(Integer, primary_key=True, autoincrement=True)
    datum = Column(Date, nullable=False)
    klient = Column(String(45), nullable=True)
    kampan = Column(String(45), nullable=True)
    cena = Column(Float, nullable=False)
    plochy = Column(JSON, nullable=False)
    
    # Relationships
    centralny_zoznam = relationship("CentralnyZoznam", back_populates="ponuka")
    klienti = relationship("Klienti", back_populates="ponuka")

class CentralnyZoznam(Base):
    __tablename__ = 'centralny_zoznam'
    idPlocha = Column(Integer, primary_key=True, autoincrement=True)
    Dodavatel = Column(String(45), nullable=False)
    Typ_plochy = Column(String(45), nullable=False)
    Region = Column(String(45), nullable=False)
    Rozmer_X_sirka = Column(Integer, nullable=False)
    Rozmer_Y_vyska = Column(Integer, nullable=False)
    Plocha_XY = Column(Float, nullable=False)
    Gps_lat = Column(Float, nullable=False)
    Gps_log = Column(Float, nullable=False)
    id_Dodavatel = Column(Integer, ForeignKey('dodavatelia.id_Dodavatel'), nullable=True)
    id_Ponuka = Column(Integer, ForeignKey('ponuka.id_Ponuka'), nullable=True)
    
    # Relationships
    dodavatel = relationship("Dodavatelia", back_populates="centralny_zoznam")
    ponuka = relationship("Ponuka", back_populates="centralny_zoznam")

class Klienti(Base):
    __tablename__ = 'klienti'
    id_klienta = Column(Integer, primary_key=True, autoincrement=True)
    mail = Column(String(45), nullable=True)
    telefon = Column(Integer, nullable=True)
    id_Ponuka = Column(Integer, ForeignKey('ponuka.id_Ponuka'), nullable=True)
    
    # Relationships
    ponuka = relationship("Ponuka", back_populates="klienti")

# Creating the engine
engine = create_engine("mysql+pymysql://root:(Mario555)@localhost/kreator_centralna_db", echo = True)

# Creating the tables in the database
Base.metadata.create_all(engine)
```

```python
def generate_google_maps_link(df, index):
  
    if index not in df.index:
        raise ValueError(f"Index {index} not found in DataFrame.")
    
    latitude = df.loc[index, 'Gps_lat']
    longitude = df.loc[index, 'Gps_log']
    
    if pd.isnull(latitude) or pd.isnull(longitude):
        raise ValueError(f"Missing GPS coordinates for index {index}.")
    
    google_maps_url = f"https://www.google.com/maps?q={latitude},{longitude}"
    return google_maps_url
```
